```python
from sklearn.cluster import KMeans
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import HDBSCAN
```

```python
import json
```

```python
file_path = "/Users/flora/graduate_school/LAB/LawCaseRetreiver/implicit_elements/larceny/factor/factors.json"

with open(file_path, 'r', encoding='utf-8') as f:
    dataset = json.load(f)
```

```python
dataset[0]
```

```python
model = SentenceTransformer('DMetaSoul/sbert-chinese-general-v2')
```

```python
vector_factors = model.encode([entry['因素'] for entry in dataset])
```

```python
NUM_CLUSTERS=20

kmeans = KMeans(n_clusters=NUM_CLUSTERS)
kmeans.fit(vector_factors)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_
print("labels: ",labels)
print("centroids: ",centroids)
```

```python
unique_labels = set(labels)
for label in unique_labels:
    print(f"Cluster {label}:")
    cluster_factors = [dataset[i] for i in range(len(labels)) if labels[i] == label]
    for factor in cluster_factors:
        print(factor['因素'])
        print(f"\t{factor['案件一']}")
        print(f"\t{factor['案件二']}")
```

```python
clustered_cases = [[] for _ in range(NUM_CLUSTERS)]
```

```python
for i, label in enumerate(unique_labels):
    cluster_factors = [dataset[i] for i in range(len(labels)) if labels[i] == label]
    for factor in cluster_factors:
        case_one = factor['案件一']
        case_two = factor['案件二']
        clustered_cases[i].append(case_one)
        clustered_cases[i].append(case_two)
```
