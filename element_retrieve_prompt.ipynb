```python
def get_response(input_content):
    completion = client.chat.completions.create(
    model=MODEL,
    messages=[
        {"role": "system", "content": "You are a helpful AI assistant built by MediaTek Research. The user you are helping speaks Traditional Chinese and comes from Taiwan."}, # <-- This is the system message that provides context to the model
        {"role": "user", "content": input_content}  # <-- This is the user message for which the model will generate a response
    ]
    )

    return completion.choices[0].message.content
```

```python
sample = origin_outlier[30:50]

response_list = []
for data in sample:
    origin = data['origin_data']['content']
    outlier = data['outlier_data']['content']
    if origin != '' and outlier != '':
        # prompt = f"我會給你一篇判決書的理由內容，理由內容為法官對「為何做出這樣的判定結果」的解釋。你的任務是根據理由內容，列出影響判決的因素並詳細說明，輸出格式請以「因素一：說明、因素二：說明...」呈現。注意，不要直接輸出人名，以被害人、被告、原告等代替。\n理由內容如下：{content}\n輸出："
        prompt = f"我會給你兩篇判決書的理由內容，理由內容為法官對「為何做出這樣的判定結果」的解釋。請將案件一與案件二分別做詳細摘要後，再依據案件一及案件二的內容差異列點列出影響不同判決的因素並詳細說明，輸出格式請以「因素一：說明、因素二：說明...」呈現。回答須包含兩部分：\n部分一：兩案的摘要\n部分二：判決不同的因素\n案件一：{origin}\n案件二：{outlier}"
        print(f"prompt: {prompt}")

        response = get_response(prompt)
    else:
        response = 'None'
    print(response)
    print(f"="*200)
    response_list.append(response)
```

```python
file_path = f"larceny/factor/30-50_content_find_factor_v1.json"

with open(file_path, 'w', encoding='utf-8') as file:
    json.dump(response_list, file, ensure_ascii=False, indent=4)
```

```python
for data in response_list:
    print(data)
    print(f"="*100)
```

```python
def extract_factors(response_list: list) -> list:
    factors = []
    for response in response_list:
        match = re.findall(r'因素(.+?)\n', response)
        cleaned_matches = [match.replace('**', '') for match in match]
        joined_factors = '；'.join(cleaned_matches)
        factors.append(joined_factors)
    return factors
```

```python
# 把每一筆的因素提出來
extracted_factors = extract_factors(response_list)
extracted_factors
```

```python
len(extracted_factors)
```

```python
factor_string = '\n'.join(extracted_factors)
print(factor_string)
```

```python
prompt = f"我會給你很多個因素，你的任務是將這些因素依照特性分群並命名。輸出格式請以「群一名字：因素一、因素二...。群二名字：因素一、因素二...」呈現。\n因素如下：{factor_string}"
classified_factor = get_response(prompt)
print(classified_factor)
```
